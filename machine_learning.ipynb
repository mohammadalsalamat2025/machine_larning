```python
import matplotlib.pyplot as plt
import numpy as np 

x_val = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y_val = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

# إنشاء نموذج متعدد الحدود من الدرجة الثالثة
mymodel = np.poly1d(np.polyfit(x_val, y_val, 3))

# طباعة معادلة النموذج
print("Polynomial Model:")
print(mymodel)

# توقع قيم جديدة باستخدام النموذج
x_new = np.linspace(1, 22, 100)
y_new = mymodel(x_new)

plt.scatter(x_val, y_val, color="red", label="Original Data")
plt.plot(x_new, y_new, color="blue", label="Fitted Curve")
plt.legend()
plt.show()
```

الكود التالي:

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))
يُستخدم لإنشاء نموذج منحنى متعدد الحدود (Polynomial Curve) لتقريب البيانات باستخدام مكتبة NumPy في لغة Python. الكود يقوم بإنشاء معادلة رياضية من الدرجة الثالثة (Polynomial of degree 3) لتناسب مجموعة من النقاط (x, y).

شرح الكود بالتفصيل:
numpy.polyfit(x, y, 3)
هذه الدالة تُستخدم لإجراء التقريب متعدد الحدود (Polynomial Fitting).
تقوم بحساب معاملات (Coefficients) معادلة متعددة الحدود التي تناسب البيانات. المعاملات الناتجة تمثل المعادلة على الشكل التالي:

حيث:
x و y: البيانات التي نحاول تقريبها.
3: درجة (Degree) كثيرة الحدود. في هذه الحالة، هي من الدرجة الثالثة.
numpy.poly1d(coefficients)
تُحوّل المعاملات الناتجة من numpy.polyfit() إلى كائن يمثل معادلة متعددة الحدود (Polynomial Function).
يسمح هذا الكائن باستخدام المعادلة بسهولة لحساب القيم أو رسم المنحنى.
كيف يعمل الكود؟
المدخلات:
المتغيرات x و y تحتوي على البيانات (قائمة من النقاط).
3 تُحدد درجة معادلة متعددة الحدود (Cubic Polynomial).
numpy.polyfit():
تحسب معاملات معادلة من الدرجة الثالثة التي تناسب البيانات.
numpy.poly1d():
تُحوّل المعاملات إلى كائن يمثل معادلة يمكن استخدامها للتنبؤ أو الرسم.
mymodel:
هو نموذج (كائن) يمكن استخدامه لتوقع القيم بناءً على معادلة منحنى التقريب.

```python
# بيانات (x, y)
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# إنشاء نموذج متعدد الحدود من الدرجة الثالثة
mymodel = np.poly1d(np.polyfit(x, y, 3))

# طباعة معادلة النموذج
print("Polynomial Model:")
print(mymodel)

# توقع قيم جديدة باستخدام النموذج
new_x = np.linspace(1, 5, 100)  # نقاط على المحور x
new_y = mymodel(new_x)          # القيم المتوقعة على المحور y

# رسم البيانات الأصلية والمنحنى
plt.scatter(x, y, color="red", label="Original Data")  # النقاط الأصلية
plt.plot(new_x, new_y, color="blue", label="Fitted Curve")  # منحنى التقريب
plt.legend()
plt.show()
```
